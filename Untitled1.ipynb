```python
train_data = tf.keras.utils.image_dataset_from_directory(
    '/content/tomatoleaf/versions/1/tomato/train',
    labels='inferred',
    label_mode='categorical',
    image_size=(256, 256),
    batch_size=32)

train_data = train_data.map(lambda x, y: (x / 255.0, y))
```

```python
val_data = tf.keras.preprocessing.image_dataset_from_directory(
    '/content/tomatoleaf/versions/1/tomato/val',
    labels='inferred',
    label_mode='categorical',
    image_size=(256, 256),
    batch_size=32)

val_data = val_data.map(lambda x, y: (x / 255.0, y))
```

```python

path = "/content/tomatoleaf/versions/1/tomato/train/Tomato___Tomato_Yellow_Leaf_Curl_Virus"


image_files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]


fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i in range(6):
    image_file = image_files[i]
    label = image_file.split('.')[0]


    img_path = os.path.join(path, image_file)
    img = mpimg.imread(img_path)
    ax = axs[i // 3, i % 3]
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(label)

plt.tight_layout()
plt.show()
```

```python
conv_base = DenseNet121(
    weights='imagenet',
    include_top = False,
    input_shape=(256,256,3),
    pooling='avg'
)
```

```python
conv_base.trainable = False
conv_base.summary()
```

```python
model = Sequential()
model.add(conv_base)
model.add(BatchNormalization())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.35))
model.add(BatchNormalization())
model.add(Dense(120, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_data, epochs=5, validation_data=val_data, callbacks=[EarlyStopping(patience=0)])
```

```python
evaluation = model.evaluate(val_data)
```

```python
model.save("plant_disease.h5")
```

```python
print("Validation Loss:", evaluation[0])
print("Validation Accuracy:", evaluation[1])
```

```python
import pickle


train_dir = "/content/tomatoleaf/versions/1/tomato/train"


class_names = sorted(os.listdir(train_dir))


class_indices = {class_name: idx for idx, class_name in enumerate(class_names)}



with open("/content/class_indices.pkl", "wb") as f:
    pickle.dump(class_indices, f)

print("Class indices created and saved successfully.")
print(class_indices)
```

```python
import cv2
import numpy as np
import tensorflow as tf

with open("/content/class_indices.pkl", "rb") as f:
    class_indices = pickle.load(f)


class_labels = {v: k for k, v in class_indices.items()}

print("[INFO] Class labels loaded:", class_labels)

def load_and_preprocess_image(image_path, target_size=(256, 256)):
    """
    Load an image from file, convert to RGB, resize, normalize, and return as a NumPy array.
    """

    image = cv2.imread(image_path)

    if image is None:
        raise ValueError("Error loading image. Please check the path.")


    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


    image = cv2.resize(image, target_size)

    image = image / 255.0
    image = np.expand_dims(image, axis=0)

    return image

def load_model(model_path):
    """
    Load a trained Keras model from the specified file path.
    """
    model = tf.keras.models.load_model(model_path)
    print("[INFO] Model loaded successfully.")
    return model

def predict_disease(model, image_path):
    """
    Predict the disease class for an input image using the trained model.
    """
    try:

        processed_image = load_and_preprocess_image(image_path)

        predictions = model.predict(processed_image)
        predicted_class_index = np.argmax(predictions, axis=1)[0]

        predicted_class_name = class_labels[predicted_class_index]

        return predicted_class_name

    except Exception as e:
        return f"Error: {e}"
```

```python

model = load_model("/content/plant_disease.h5")  # Change this to your actual model path

image_path = "/content/tomatoleaf/versions/1/tomato/val/Tomato___Target_Spot/01b2013e-4030-4cd0-843c-2dbacf5f3758___Com.G_TgS_FL 8398.JPG"  # Change this to the actual image path

predicted_disease = predict_disease(model, image_path)
print(f"Predicted Disease: {predicted_disease}")
```
